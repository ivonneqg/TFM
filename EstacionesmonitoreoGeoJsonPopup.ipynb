```
ruta_dbfs = "dbfs:/FileStore/shared_uploads/ivonneqg@outlook.com/Estaciones_PM10.csv"

estaciones = spark.read\
            .option("encoding", "Cp1252")\
            .option("header", "true")\
            .option("delimiter", "¥")\
            .option("inferSchema", "true")\
            .csv(ruta_dbfs)

estaciones.show(5)
```

```
#estaciones['COORDENADAS'] = estaciones.select('COORDENADAS').replace('""', '"')
df = estaciones.toPandas()



df.head()
#df.dtypes
```

```
import folium
import json
import geojson
from folium.plugins import HeatMap

m=folium.Map([4.660791, -74.151521], tiles='stamentoner', zoom_start=12, width='70%', height='70%')

longs = []
lats = []

for ind in df.index:
    v_geojson=geojson.loads(json.dumps(json.loads(df['COORDENADAS'][ind])))
    
    # extrayendo los valores de longitud y latitud a listas separadas
    longs.append(v_geojson.coordinates[1])
    lats.append(v_geojson.coordinates[0])
    
    folium.Marker( v_geojson.coordinates
                  , popup=(
                      '#'+(df['STA_SERIALCODE'][ind]).astype(str) + ' - ' + (df['STA_STATIONNAME'][ind])
                 )).add_to(m)
    

# calcular los valores medios de longitud y latitud
import statistics
meanLong = statistics.mean(longs)
meanLat = statistics.mean(lats)
# crear un objeto mapa base usando Map()
##mapObj = folium.Map(location=[meanLat, meanLong], zoom_start = 14.5)
# crear capa de mapa de calor
heatmap = HeatMap( list(zip(lats, longs, df['MON_NUMERICFORMAT'])),
                   min_opacity=0.2,
                   max_val=df['MON_NUMERICFORMAT'].max(),
                   radius=50, blur=50, 
                   max_zoom=1)
# añadir capa de mapa de calor al mapa base
heatmap.add_to(m)

m
```
