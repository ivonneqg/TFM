```
ruta_dbfs = "dbfs:/FileStore/shared_uploads/ivonneqg@outlook.com/EstacionesMonitoreo_2.csv"
estaciones = spark.read\
            .option("encoding", "Cp1252")\
            .option("header", "true")\
            .option("delimiter", "¥")\
            .option("inferSchema", "true")\
            .csv(ruta_dbfs)

estaciones.show()
```

```
df = estaciones.toPandas()
df.head()
```

```
#estaciones['COORDENADAS'] = estaciones.select('COORDENADAS').replace('""', '"')
#df = estaciones.toPandas()

for ind in df.index:
    df['COORDENADAS'][ind] = df['COORDENADAS'][ind].replace('""', '"')
    df['COORDENADAS'][ind] = df['COORDENADAS'][ind].replace('"{', '{')
    df['COORDENADAS'][ind] = df['COORDENADAS'][ind].replace('}"', '}')

df.head()
#df.dtypes
```

```
import folium
import json
import geojson

m=folium.Map([4.660791, -74.151521], zoom_start=12, width='70%', height='70%')

for ind in df.index:
    v_geojson=geojson.loads(json.dumps(json.loads(df['COORDENADAS'][ind])))
    #print(v_geojson)
    ag = folium.GeoJson(v_geojson)
    ag.add_child(folium.Popup(
                      '#' + (df['STA_STATIONCODE'][ind]).astype(str) 
                 ))
    ag.add_to(m)
    
    folium.Marker( v_geojson.coordinates
                  , popup=(
                      '#' + (df['STA_STATIONCODE'][ind]).astype(str)
                 )).add_to(m)

m
```

```
SELECT  S.*  '{"type":"Point","coordinates":[' || S.STA_LATITUDE || ',' || S.STA_LONGITUDE || ']}' COORDENADAS
FROM RMCAB.TB_STATION S
WHERE S.STA_LATITUDE IS NOT NULL
AND S.STA_LONGITUDE IS NOT NULL
AND S.STA_LATITUDE != '0'
AND S.STA_LONGITUDE != '0';
```
